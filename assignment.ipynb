```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df= pd.read_csv('land_prices.csv')
print(df.head)
```

```python
df['price_per_cent']=df['price_lakhs']/df['land_area_cents']
print(df.columns)
```

```python
numeric_feature=['land_area_cents', 'distance_to_school_km',
       'distance_to_airport_km', 'distance_to_railway_station_km',
       'distance_to_hospital_km', 'distance_to_medical_college_km',
       'distance_to_bus_stop_km', 'distance_to_market_km']

df_processed= pd.get_dummies(df,columns=['location_name', 'taluk', 'village', 'land_type'],drop_first=True)
print(df_processed.columns)
```

```python
dummy_features=[]
for columns in df_processed.columns:
  if columns.startswith('location_name') or columns.startswith('taluk') or columns.startswith('village') or columns.startswith('land_type'):
    dummy_features.append(columns)
print(dummy_features)
```

```python
features=numeric_feature+dummy_features
print(features)
```

```python
x=df_processed[features]
y=df_processed['price_per_cent']

x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
comparison=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
print(comparison)
```

```python
features_list=model.feature_names_in_.tolist()
print(features_list)
```
